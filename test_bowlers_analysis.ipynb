#### Dataset: Bowling records in test matches dataset that include specific features such as

1. **Player**: Name of the individual player followed by his country name/initials in parenthesis

2. **Span**: Starting and end year of the player's test career

3. **Mat**: Number of test matches played by the player

4. **Inns**: Number of innings of test matches bowled by the player

5. **Balls**: Number of balls bowled by the player

6. **Runs**: Number of runs conceded by the player

7. **Wkts**: Number of wickets taken by the player

8. **BBI**: Best bowling figure of the player within an innings

9. **BBM**: Best bowling figure of the player within a test match

10. **Ave**: Bowling average of the player (the number of runs conceded per wicket taken)

11. **Econ**: Bowling economy of the player (the average number of runs they have conceded per over bowled)

12. **SR**: Bowling strike rate of the player (the average number of balls bowled per wicket taken)

13. **5**: No of innings the player took 5 wicktes

14. **10**: No of matches the player took 10 wicktes

#### Renaming the column names to make the dataset more user friendly

```python
df = df.rename(columns={'Player':'Player_Name',
                        'Span' : 'Career_duration',
                        'Mat':'Matches_playes', 
                        'Inns':'Innings_bowled',
                        'Balls': 'Balls_bowled',
                        'Runs': 'Runs_conceded',
                        'Wkts' : 'Wickets_taken',
                        'BBI' : 'Best_innings_bowling',
                        'BBM' : 'Best_match_bowling',
                        'Ave': 'Bowling_average',
                        'Econ' : 'Bowling_economy',
                        'SR' : 'Bowling_strike_rate',
                        '5' : '5_wickets_in_an_innings',
                        '10' : '10_wickets_in_a_match'})

display(df.head())
```

#### Removing the columns BBI and BBM from the dataframe

```python
df.drop(['Best_innings_bowling', 'Best_match_bowling'], axis=1, inplace=True)

display(df.head())
```

#### Finding out the number of players who played for ICC

```python
# For this first, we have to split the "Player_Name" column into two different columns one will have name only and other team/country affiliation
df_player = df['Player_Name'].str.split("(", expand=True)
display(df_player.head(10))
```

```python
# Second we have to concatenate this new dataframe to previous dataframe
df = pd.concat([df, df_player], axis=1)
# Propoerly naming the newly added columns
df = df.rename(columns={0: 'Player',
                        1: 'Country'})

display(df.head())
```

**Comment**: Looking at the dataframe we have to discard the redundant column **Player_Name**, have to properly order the data columns, **Country** column needs further attention to cleaning and extracting information for ICC affiliation

```python
# Removing Player_Name column 
df.drop('Player_Name', axis=1, inplace=True)
# Looking for existing column names
print(df.columns)
```

```python
# Assinging preferred column sequence
new_col_sequence = ['Player', 'Country', 'Career_duration', 'Matches_playes', 'Innings_bowled', 'Balls_bowled',
                    'Runs_conceded', 'Wickets_taken', 'Bowling_average', 'Bowling_economy', 'Bowling_strike_rate',
                    '5_wickets_in_an_innings', '10_wickets_in_a_match']
df = df[new_col_sequence]

display(df.head())
```

```python
#Cleaning the Country column
df['Country'] = df['Country'].str.replace(")", "")
display(df.head())
```

```python
# Defining a function to find out ICC affiliation of each player
def icc_check(x):
    if "ICC" in x:
        return "Yes"
    else:
        return "No"
```

```python
# Applying the defined function **icc_check** on **Country** coulmn
df['ICC_affiliation'] = df['Country'].apply(icc_check)

display(df.head(10))
```

```python
df['ICC_affiliation'].value_counts()
```

**Comment**: The above output indicates there are only 5 players out of 79 in this test bowlers dataframe played for ICC

#### Finding out which player(s) had played for the longest period of time and who had played for the shortest period of time

```python
# First we have to split the 'Carrer_duration' column based on the "-"
df_car_dur = df['Career_duration'].str.split("-", expand=True)

# Second we need to concatinate the new dataframe with the main dataframe
df = pd.concat([df, df_car_dur], axis=1)
display(df.head())
```

```python
# Third we have to rename the newly created column names
df = df.rename(columns={0: "start_year",
                        1: "end_year"})

# As the Career_duration now becomes redundant we need to remove it
df = df.drop('Career_duration', axis=1)

display(df.head())
```

```python
# Quickly check for data type of newly created columns
df.info()
```

**Comment**: As the start_year and end_year are considered object we need to transform them into integers

```python
# Converting year columns to integer
df['start_year'] = df['start_year'].astype('int') 
df['end_year'] = df['end_year'].astype('int')

print(df.info())
```

```python
# Creating a new column Number of years
df['number_of_years'] = df['end_year'] - df['start_year']

# Assinging preferred column sequence
print(df.columns)
```

```python
new_col_sequence2 = ['Player','number_of_years','Country', 'Matches_playes', 'Innings_bowled', 'Balls_bowled',
                     'Runs_conceded', 'Wickets_taken', 'Bowling_average', 'Bowling_economy', 'Bowling_strike_rate',
                     '5_wickets_in_an_innings', '10_wickets_in_a_match', 'ICC_affiliation', 'start_year','end_year' ]
df = df[new_col_sequence2]
display(df.head(10))
```

```python
df.sort_values(by='number_of_years', ascending = False).head(5)
```

**Comment**: The above output informs **Imran Khan** had played for the longest period of time which is 21 years.

```python
df.sort_values(by='number_of_years', ascending = True).head(5)
```
